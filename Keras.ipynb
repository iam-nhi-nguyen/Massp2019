## Keras trong Linear regression

```
import numpy as np # cho tính toán
from keras.models import Sequential # cho xây dựng sequential model
from keras.layers.core import Dense, Activation # cho các lớp neuron và cho các hàm activation
from keras import optimizers # cho thuật toán tối ưu

# 1. create pseudo data y = 3*X + 4
X = np.random.rand(1000, 1) # chọn vecto X input có 1000 chiều ngẫu nhiên
y = 4 + 3 * X + .2*np.random.randn(1000, 1) # noise added

# 2. Build model 
model = Sequential([Dense(1, input_shape = (1,), activation='linear')]) # 1 unit ở lớp output; 1 lớp input là 1 vecto 1 chiều; activation là công thức linear, 

# 3. gradient descent optimizer and loss function 
sgd = optimizers.SGD(lr=0.1) # optimizer là SGD; learning rate = 0.1
model.compile(loss='mse', optimizer=sgd) # loss function là mean squared error, optimizer là SGD

# 4. Train the model 
model.fit(X, y, epochs=100, batch_size=2) # giới hạn vòng lặp = 100, kích thước mini-batch = 2

# 5. Xem hệ số
w = model.get_weights()
```

```
import matplotlib.pyplot as plt # cho vẽ đồ thị

# Display result
w = model.get_weights()
w_0 = w[1]
w_1 = w[0][0]
x0 = np.linspace(0, 1, 2, endpoint=True) # khoảng input vẽ trên đồ thị
y0 = w_0 + w_1*x0

# Draw the fitting line 
plt.plot(X.T, y.T, 'b.')     # data 
plt.plot(x0, y0, 'y', linewidth = 2)   # the fitting line
plt.axis([0, 1, 0, 10]) # khoảng giá trị trên các trục
plt.show()
```
